```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
teamsCsv = pd.read_csv("mlb_teams.csv", encoding='utf-16')
gamesCsv = pd.read_csv("mlb_games.csv", encoding='utf-16')
```

```python
teams = teamsCsv['prefix_1'].values.tolist()
print(teams)
```

```python
games = gamesCsv[["home_team","away_team"]].values.tolist()
print(games)
```

```python
#3-Ynsz bir a oluturucak ve tekrar eden kenarlar silinecek
G = nx.Graph()
G.add_nodes_from(teams)
G.add_edges_from(games)
G.number_of_nodes()
```

```python
#4-Network izdilecek(network ismi G)
nx.draw(G, with_labels=True)
plt.show()
```

```python
#5-Ynl Network oluturulacak ve izdilecek(network ismi G_directed)
G_directed = nx.DiGraph()
G_directed.add_nodes_from(teams)
G_directed.add_edges_from(games)
```

```python
nx.draw(G_directed, with_labels=True)
plt.show()
```

```python
#6-G network iin tm dmlerin dereceleri listelenecek
print(dict(G.degree()))
```

```python
#7-G network iin ortalama derece hesaplanacak
avg_degree = sum(dict(G.degree()).values()) / G.number_of_nodes()
print(avg_degree)
```

```python
#8-G network iin histogram izdirilecek
degrees = [d for n, d in G.degree()]

plt.hist(degrees, bins=range(min(degrees), max(degrees)+2))
plt.xlabel("Derece")
plt.ylabel("Frekans")
plt.title("Derece Histogram")
plt.show()
```

```python
#9-G network iin her dmn Degree centrality deerleri hesaplanacak
degree_c = nx.degree_centrality(G)
print(degree_c)
```

```python
#10-G network iin her dmn Closeness centrality deerleri hesaplanacak
closeness = nx.closeness_centrality(G)
print(closeness)
```

```python
#11- G network iin her dmn Betweenness centrality deerleri hesaplanacak
bet_c = nx.betweenness_centrality(G)
print(bet_c)
```

```python
#12- G network iin her dmn Eigenvector centrality deerleri hesaplanacak
eig = nx.eigenvector_centrality(G)
print(eig)
```

```python
#13-G network iin her dmn Katz centrality deerleri hesaplanacak
katz = nx.katz_centrality(G, alpha=0.01, beta=1.0, max_iter=10000)
print(katz)
```

```python
#14-G network iin her dmn PageRank centrality deerleri hesaplanacak
pr = nx.pagerank(G)
print(pr)
```

```python
#15-G network iin modularity hesaplanacak
from networkx.algorithms.community import greedy_modularity_communities, modularity

communities = greedy_modularity_communities(G)
print(len(communities))

mod = modularity(G, communities)
print(mod)
```

```python
#16- G network iin density hesaplanacak
dens = nx.density(G)
print(dens)
```

```python
#17- G network iin average clustering hesaplanacak
avg_clust = nx.average_clustering(G)
print(avg_clust)
```

```python
#18- G network iin diameter hesaplanacak
diam = nx.diameter(G)
print(diam)
```
